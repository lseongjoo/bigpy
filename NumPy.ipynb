마방진(魔方陣)은 n<sup>2</sup>개의 수를 가로, 세로, 대각선 방향의 수를 더하면 모두 같은 값이 나오도록 n × n 행렬에 배열한 것이다. 마법진(魔法陣), 낙서(洛書)라고도 한다. 일반적으로 마방진의 각 칸에는 1부터 n2까지의 수가 한 개씩 들어간다. 마방진은 n이 2일 때를 제외하고 항상 존재한다. (위키백과, https://goo.gl/fWHxg0)

M(n) = (n<sup>3</sup> + n) / 2

3x3 마방진 --> M(3) = 15

## 순수 파이썬

단일 종목의 90일 가격 시뮬레이션

```python
import random
prices = [1000]
changes = [np.random.uniform(-0.3, 0.3) for i in range(90)]
for delta in changes:
    prices.append(prices[-1]*(1+delta))
prices[:3]
```

10개 종목의 90일 가격 시뮬레이션

```python
import random
prices = [[np.random.randint(5,1000)] for i in range(10)]
changes = [[np.random.uniform(-0.3, 0.3) for i in range(90)] for i in range(10)]
for p, c in zip(prices, changes):
    for delta in c:
        p.append(p[-1]*(1+delta))

prices[0][:3]
```

## NumPy로 수행

단일 종목의 90일 가격 시뮬레이션

```python
changes = 1 + np.random.uniform(low=-0.3, high=0.3, size=90)
prices = 1000 * changes.cumprod()
prices[:5]
```

10개 종목의 90일 가격 시뮬레이션

```python
import numpy as np
prices = np.random.randint(5,100, size=10)
changes = 1+np.random.uniform(low=-0.3, high=0.3, size=(90,10))
prices =  prices * changes.cumprod(0)
print(prices.shape)
prices[:3,:3]
```
